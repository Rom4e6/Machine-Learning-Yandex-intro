```python
import pandas
data = pandas.read_csv('titanic.csv', index_col='PassengerId')
data[:0]
```

```python
people = data['Sex'].value_counts()
#men = data['Sex'].value_counts('male')
#women = data['Sex'].value_counts('female')
print(people.male, people.female, sep=' ')
ans_1= open('1.txt', 'w')
ans = str(people.male)+' '+str(people.female)
ans_1.write(ans)
ans_1.close()
```

```python
#Какой части пассажиров удалось выжить? Посчитайте долю выживших пассажиров
survived = data['Survived'].value_counts()
serv_count=round(survived[1]/(survived[0]+survived[1])*100,2)
print(serv_count)
ans_2 = open('2.txt', 'w')
ans_2.write(str(serv_count))
ans_2.close()
```

```python
#Какую долю пассажиры первого класса составляли среди всех пассажиров? 
pclass = data['Pclass'].value_counts()
first_class = round(pclass[1]/pclass.sum()*100,2)
#print(pclass)
print(first_class)
ans_3 = open('3.txt', 'w')
ans_3.write(str(first_class))
ans_3.close()
```

```python
#Какого возраста были пассажиры? Посчитайте среднее и медиану возраста пассажиров.
#В качестве ответа приведите два числа через пробел
age = data['Age']
age_mean=round(age.mean(),2)
age_median=age.median()
print(age_mean)
print(age_median)
ans_4= open('4.txt', 'w')
ans = str(age_mean)+' '+str(age_median)
ans_4.write(ans)
ans_4.close()
```

```python
#Коррелируют ли число братьев/сестер/супругов с числом родителей/детей? 
#Посчитайте корреляцию Пирсона между признаками SibSp и Parch.
d = data[['SibSp', 'Parch']]
df = pandas.DataFrame(d,  columns = ['Parch','SibSp']) #создаём датафрейм для сравнения
df.corr(method='pearson')
```

```python
#Оставьте в выборке четыре признака: класс пассажира (Pclass), цену билета (Fare), возраст пассажира (Age) и его пол (Sex).
d = data[['Pclass', 'Fare', 'Age','Sex', 'Survived']]
df = pandas.DataFrame(d,  columns = ['Pclass', 'Fare', 'Age','Sex','Survived']) #создаём датафрейм для сравнения
```

```python
#Выделите целевую переменную — она записана в столбце Survived.
```

```python
#В данных есть пропущенные значения — например, для некоторых пассажиров неизвестен их возраст. 
#Такие записи при чтении их в pandas принимают значение nan. 
#Найдите все объекты, у которых есть пропущенные признаки, и удалите их из выборки.
df=df[['Pclass', 'Fare', 'Age','Sex', 'Survived']].dropna()
```

```python
import numpy 
from sklearn.tree import DecisionTreeClassifier
```

```python
#Обратите внимание, что признак Sex имеет строковые значения.
filtered_df = df[["Pclass", "Fare", "Age", "Sex"]].replace("female", 0).replace("male", 1) #теперь нет
```

```python
#Обучите решающее дерево с параметром random_state=241 и остальными параметрами по умолчанию 
#(речь идет о параметрах конструктора DecisionTreeСlassifier).
classifier = DecisionTreeClassifier(random_state=241)
X = filtered_df #множество входных данных с предобработкой
Y = df['Survived'] # соответсвующеее ему множестов ответов
classifier.fit(X,Y) #обучаем наше дерево
importances = classifier.feature_importances_
print(importances)
print(df)
```
