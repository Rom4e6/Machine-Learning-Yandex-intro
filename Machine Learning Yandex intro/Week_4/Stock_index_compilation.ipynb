```python
import pandas as pd
from sklearn.decomposition import PCA

file = pd.read_csv('close_prices.csv', index_col = 'date')
df = pd.DataFrame(data = file)
pca = PCA(n_components = 10)
pca.fit(df)
```

```python
components_amount = 0 
summ = 0
for i in pca.explained_variance_ratio_:
    summ+= i
    components_amount+=1
    if summ>.9:
        break
print(components_amount)
```

```python
df = pca.transform(df)
```
